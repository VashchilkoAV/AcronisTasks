```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("ms_bucket.csv")
```

```python
#удаляем строки с Nan
print(data.shape)
data = data.dropna()
print(data.shape)
```

```python
#конвертируем  timestamp в date+time
data['timestamp'] = data.apply(lambda x: pd.Timestamp(x['timestamp'], unit='s'), axis=1)
```

```python
data.head(5)
```

```python
cols = ['start', 'end'] + (data.columns.to_list()[1:])
```

```python


processedData = pd.DataFrame(columns=cols)
processedData.head(5)
```

```python
#Преобразуем датасет так, чтобы строки отражали запросы, обработанные за промежуток времени от i до i+1 строки
print(data.shape[0]-1)
for i in range(data.shape[0]-1):
    processedData.loc[i] = [data.iloc[i, 0]] + [data.iloc[i+1, 0]] + [data.iloc[i+1, j]-data.iloc[i, j] for j in range(1, data.shape[1])]
    #print([data.iloc[i, 0]] + [data.iloc[i+1, 0]] + [data.iloc[i+1, j]-data.iloc[i, j] for j in range(1, data.shape[1]-1)])
```

```python
processedData[:]
```

```python
#отсортируем колонки в порядке возрастания
cols2 = processedData.columns.to_list()[3:]
cols2pr = [int(i) for i in cols2]
cols2pr.sort()
cols2pr
cols2prStr = [str(i) for i in cols2pr]
cols2prStr = ['start', 'end']+ cols2prStr+['Inf']
processedData = processedData[cols2prStr]
processedData[:]
```

```python
#Изменим датасет так, чтобы колонки показывали количество запросов, обработанных между i и i+1 колонками
cols2 = ['start', 'end', 'total', 'less than 1'] + [cols2prStr[i]+'-'+cols2prStr[i+1] for i in range(2, len(cols2prStr)-1)] 
print(cols2)
finalData = pd.DataFrame(columns=cols2)

print(finalData.shape)

for i in range(processedData.shape[0]):
    finalData.loc[i] = [processedData.iloc[i, 0]] + [processedData.iloc[i, 1]] + [processedData.iloc[i, processedData.shape[1]-1]] + [processedData.iloc[i, 2]] + [processedData.iloc[i, j+1]-processedData.iloc[i, j] for j in range(2, processedData.shape[1]-1)]
    #print([processedData.iloc[i, 0]] + [processedData.iloc[i, 1]] +[processedData.iloc[i, 2]] + [processedData.iloc[i, j+1]-processedData.iloc[i, j] for j in range(3, processedData.shape[1]-1)])
```

```python
finalData[:]
```

```python
#Вычислил средние значения времени выполнения запроса в зависимости от дня/времени
#print(cols2pr)
def EvaluateAvg(arr) :
    num = sum(arr)
    S = 0
    for i in range(len(arr)):
        S += arr[i]*cols2pr[i];
    return S/num

avgs = [EvaluateAvg(finalData.iloc[i, 3:-1].to_list()) for i in range(finalData.shape[0])]
#avgs
```

```python
print(avgs)
```

```python
plt.plot([finalData.iloc[i,0] for i in range(len(avgs))], avgs)
```

```python
#Вычислил дисперсии времени выполнения запроса в зависимости от дня/времени
def EvaluateVar(arr, avg):
    ret = 0
    for i in range(len(arr)):
        
        ret += (cols2pr[i]-avg)*(cols2pr[i]-avg)*arr[i]
    return ret

disps = [EvaluateVar(finalData.iloc[i, 3:-1].to_list(), avgs[i]) for i in range(finalData.shape[0])]
```

```python
print(disps)
```

```python
plt.plot([finalData.iloc[i,0] for i in range(len(avgs))], disps)
```
