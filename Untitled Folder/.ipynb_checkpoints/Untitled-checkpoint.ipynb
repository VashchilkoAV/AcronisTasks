```python
#сформируем обучающую и тестовую выборки
X = data.drop(['Unnamed: 0', 'tag'], 1)
y = data['tag']



from sklearn.model_selection import train_test_split


(X_train, 
 X_test, 
 y_train, y_test) = train_test_split(X, y, 
                                     test_size=0.3, 
                                     random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
```

```python
est = LogisticRegression()
param_grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10]}
cv = 3

opt = GridSearchCV(est, param_grid, cv = cv)
opt.fit(X_train, y_train)
```

```python
auc1 = roc_auc_score(y_test, opt.predict_proba(X_test)[:,1])
```

```python
print(auc1)
```

```python
#with stratification
(X_train, 
 X_test, 
 y_train, y_test) = train_test_split(X, y, 
                                     test_size=0.3, 
                                     random_state=0, stratify=y)
```

```python
est = LogisticRegression()
param_grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10]}
cv = 3

opt = GridSearchCV(est, param_grid, cv = cv)
opt.fit(X_train, y_train)
```

```python
auc1 = roc_auc_score(y_test, opt.predict_proba(X_test)[:,1])
print(auc1)
```

```python
#Используем отбор признаков с помощью Лассо
est = LogisticRegression(class_weight='balanced', penalty='l1')
param_grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10]}
cv = 3

opt = GridSearchCV(est, param_grid, cv = cv)
opt.fit(X_train, y_train)

auc1 = roc_auc_score(y_test, opt.predict_proba(X_test)[:,1])
print(auc1)
```

```python
#Лассо оказался наиболее эффективным
print(opt.best_estimator_.coef_[0][:])
```
